# Getting started

Once you've chosen your scenario, download the data from [the Iowa website](https://data.iowa.gov/Economy/Iowa-Liquor-Sales/m3tr-qhgy) in csv format. Start by loading the data with pandas. You may need to parse the date columns appropriately.

```python
import pandas as pd

## Load the data into a DataFrame
ldata = pd.read_csv("Iowa_Liquor_sales_sample_10pct.csv")
## Transform the dates if needed, e.g.
ldata["Date"] = pd.to_datetime(ldata["Date"], format="%m/%d/%Y")
ldata.head(5)
```

```python
ldata.columns
```

```python
type(ldata)
```

```python
ldata = ldata.dropna(how='any')

ldata.head(5)
```

```python
type(ldata)
```

```python
type(ldata['State Bottle Cost'])
```

```python
ldata['State Bottle Cost']  = ldata['State Bottle Cost'].map(lambda x: x.lstrip('$'))
```

```python
ldata['State Bottle Retail'] = ldata['State Bottle Retail'].map(lambda x: x.lstrip('$'))
```

```python
ldata['Sale (Dollars)'] = ldata['Sale (Dollars)'].map(lambda x: x.lstrip('$'))
```

```python
ldata['State Bottle Cost'] = ldata['State Bottle Cost'].convert_objects(convert_numeric=True)
```

```python
ldata['State Bottle Retail'] = ldata['State Bottle Retail'].convert_objects(convert_numeric=True)
```

```python
ldata['Sale (Dollars)'] = ldata['Sale (Dollars)'].convert_objects(convert_numeric=True)
```

```python
ldata['Zip Code'] = ldata['Zip Code'].convert_objects(convert_numeric=True)
```

```python
ldata.describe()
```

```python
pivot_liq = pd.pivot_table(ldata, index = ['City'], aggfunc = sum).sort(['Volume Sold (Liters)'])
#
#pd.set_option('precision', 0)
```

```python
most_drinking  = pivot_liq.iloc[-30:,[1,5,10,11]] # We take 30 most drinking cities
most_drinking
```

```python
cities = most_drinking.index.values.tolist() #list of most drinking cities
```

```python
most_drinking['Bottle Size'] = (most_drinking['Volume Sold (Liters)']/most_drinking['Bottles Sold'])
most_drinking['Price per Litre'] = (most_drinking['Volume Sold (Liters)']/most_drinking['Sale (Dollars)'])
pd.set_option('precision', 1)
most_drinking
```

```python
most_drinking.describe()
```

```python
result = pd.pivot_table(ldata, index=['City'])
result['Price per Litre'] = (result['Volume Sold (Liters)']/result['Sale (Dollars)'])
result.describe() 
```
